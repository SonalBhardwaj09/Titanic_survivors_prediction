```python

# Assuming df is your original DataFrame

# Define the conditions
condition_male = ndf['Sex'] == 'male'
condition_age_less_than_10 = ndf['Age'] < 10 & 40
condition_class_1 = ndf['Pclass'] == 1
condition_embarked_s = ndf['Embarked'] == 'S'

# Create a new column with selected values based on conditions
ndf['Survived'] = np.select(
    [condition_male & condition_age_less_than_10,
     condition_male & ~condition_age_less_than_10 & condition_class_1,
     condition_embarked_s],
    [0, 1, 1],  # Corresponding values for each condition
    default=0  # Default value for other cases
)

# Create a new DataFrame with specified columns
new_df = ndf[['PassengerId', 'Survived']]

# Display the resulting DataFrame
print(new_df)
```

```python
new_df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Assuming df is your DataFrame
# You can replace it with the actual name of your DataFrame

# Select features and target variable
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
target = 'Survived'

# Create a DataFrame with only the selected features and target variable
data = ndf[features + [target]]

# Drop rows with missing values
data = data.dropna()

# Convert categorical variables to numerical using one-hot encoding
data = pd.get_dummies(data, columns=['Sex', 'Embarked'], drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(target, axis=1),
    data[target],
    test_size=0.2,
    random_state=42
)

# Create and train the Random Forest classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```
